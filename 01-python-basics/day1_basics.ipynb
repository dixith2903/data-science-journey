## Simple Calculator

```python
x = float(input("Enter first number: "))
y = float(input("Enter second number: "))

print("Sum:", x + y)
print("Difference:", x - y)
print("Product:", x * y)
print("Division:", x / y)
```

## Unit Converter (KM to Meters)

```python
km = float(input("Enter kilometers: "))
meters = km * 1000
print("Miles:", meters)
```
